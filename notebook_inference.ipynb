```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

#OLS
import statsmodels.api as sm

#Linearmodel
from sklearn import linear_model
```

```python
#import the dataset
df_californie = pd.read_csv('model_ML/df_californie.csv')
df_californie
```

```python
# target variable
x= df_californie.drop('median_house_value', axis=1)
y= df_californie['median_house_value']

#import statsmodels prediction
from statsmodels.sandbox.regression.predstd import wls_prediction_std

#plot the statsmodels
x1= sm.add_constant(x)
model = sm.OLS(y, x1)
results = model.fit()

#show
print(results.summary())
```

**R-Squared :** R2 est une statistique qui donnera des informations sur la qualité de l'ajustement d'un modèle. Il varie de 0 à 1.  

**Dans notre cas, la valeur du R au carré est de 0,636, ce qui explique que 63% de la variance est expliquée par le modèle.**  

**Statistiques F et Prob Statistiques F :** Les statistiques « F value » et « Prob(F) » testent la signification globale du modèle de régression. Plus précisément, ils testent l'hypothèse nulle selon laquelle tous les coefficients de régression sont égaux à zéro. La valeur F est le rapport de la somme des carrés de la régression moyenne divisée par la somme des carrés de l'erreur moyenne. Sa valeur va de zéro à un nombre arbitrairement grand.  

**AIC & BIC :** *AIC* est l'abréviation de *Akaike's Information Criteria* et est utilisé pour la sélection du modèle. Il pénalise le mode erreurs dans le cas où une nouvelle variable est ajoutée à l'équation de régression. Il est calculé comme le nombre de paramètres moins la vraisemblance du modèle global. Un AIC inférieur implique un meilleur modèle.  
Alors que *BIC* signifie critères *d'information bayésiens* et est une variante de l'AIC où les sanctions sont rendues plus sévères.

**coef :** calcule du coéficient. 

**std err :** Il montre la précision pour chaque prédiction. Réduisez l'erreur standard donne de meilleures estimations.  

**t & P>|t| :** affiche la valeur des statistiques **t** et la valeur **p**. Cela implique des hypothèses. Cela répond à la question est-ce une variable utile ou nous aide-t-elle à expliquer la variabilité que nous avons dans ce cas. Comme nous le savons, une valeur p < 0,05 est considérée comme une variable significative. Et nous pouvons dire dans notre exemple que la "taille" est un prédicteur important tout en prédisant le "prix".  

**Omnibus/Prob(Omnibus) :** Un test omnibus est effectué afin de vérifier si les erreurs sont normalement distribuées (l'une des hypothèses de la régression linéaire).  

**Skew :** La valeur de Skew est préférée proche de zéro, indiquant que la distribution résiduelle est normale. Notez que cette valeur contrôle également l'Omnibus.  

**Kurtosis :** C'est une mesure de la courbure (point culminant) des données. Des pics plus élevés conduisent à une plus grande kurtose. Une valeur plus élevée pour Kurtosis montre un regroupement plus serré des résidus autour de zéro, ce qui implique un meilleur modèle avec peu de valeurs aberrantes.  

**Durbin-Watson :** Il teste l'homoscédasticité (indépendance de l'erreur). Une valeur entre 1 et 2 est préférée.  

**Jarque-Bera (JB)/Prob(JB) :** Ce test concerne la normalité du résidu (l'une des hypothèses de régression linéaire) Le test porte le nom de Carlos Jarque et Anil K. Bera. La statistique du test est toujours positive. Une grande valeur du test de Jarque-Bera montre que les erreurs ne sont pas normalement distribuées.  

**Numéro de condition :** Ce test mesure la sensibilité de la sortie d'une fonction par rapport à son entrée. En cas de multicolinéarité, on peut s'attendre à des fluctuations beaucoup plus élevées à de petits changements dans les données.
